```python
from sklearn.cluster import KMeans
import pandas
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot
%matplotlib inline
```

```python
data_frame = pandas.read_csv("osu!_beatmaps_data.csv", encoding="iso-8859-1")
data_frame.head()
```

```python
pyplot.scatter(data_frame.length, data_frame.count_normal)
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(data_frame[['count_normal', 'length']])
y_predicted
```

```python
data_frame['cluster'] = y_predicted
data_frame.head()
```

```python
data_frame_1 = data_frame[data_frame.cluster == 0]
data_frame_2 = data_frame[data_frame.cluster == 1]
data_frame_3 = data_frame[data_frame.cluster == 2]
pyplot.scatter(data_frame_1.length, data_frame_1.count_normal)
pyplot.scatter(data_frame_2.length, data_frame_2.count_normal)
pyplot.scatter(data_frame_3.length, data_frame_3.count_normal)

pyplot.xlabel('Length')
pyplot.ylabel('Count normal')
```

```python
scaler = MinMaxScaler()
scaler.fit(data_frame[['length']])
data_frame.length = scaler.transform(data_frame[['length']])

scaler.fit(data_frame[['count_normal']])
data_frame.count_normal = scaler.transform(data_frame[['count_normal']])

scaler.fit(data_frame[['stars']])
data_frame.stars = scaler.transform(data_frame[['stars']])

scaler.fit(data_frame[['stack_leniency']])
data_frame.stack_leniency = scaler.transform(data_frame[['stack_leniency']])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(data_frame[['count_normal', 'length', 'stars', 'stack_leniency']])
```

```python
data_frame['cluster'] = y_predicted
data_frame.head()
```

```python
data_frame_1 = data_frame[data_frame.cluster == 0]
data_frame_2 = data_frame[data_frame.cluster == 1]
data_frame_3 = data_frame[data_frame.cluster == 2]
pyplot.scatter(data_frame_1.length, data_frame_1.count_normal)
pyplot.scatter(data_frame_2.length, data_frame_2.count_normal)
pyplot.scatter(data_frame_3.length, data_frame_3.count_normal)

pyplot.xlabel('Length')
pyplot.ylabel('Count normal')
```

```python
k_range = range(1, 15)
sum_square_err = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(data_frame[['count_normal', 'length', 'stars', 'stack_leniency']])
    sum_square_err.append(km.inertia_)
```

```python
sum_square_err
```

```python
pyplot.plot(k_range, sum_square_err)
```
